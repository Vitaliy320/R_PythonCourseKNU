The marketing campaign dataset contains infromation about customers who make thier purchases on shops or in the internet.
The dataset's columns describe a custmer's birth date, marital status, number of children (non-teenagers and teenagers), amount of days since the last transaction (recency), amount of money spent on a particular category of goods (MntWines - MntGoldProds), being involved into one of marketing campaings (AcceptedCmp1 - AcceptedCmp5) etc.

The dataset is designed to get insights about different types of clients in order to create more effective marketing campaigns in future.

```python
import pandas as pd

data = pd.read_csv('marketing_campaign.csv', sep="\t")
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'], dayfirst=True)
data['Spending'] = data['MntWines'] + data['MntFruits']+ data['MntFishProducts']+ data['MntSweetProducts']+ data['MntGoldProds']
data
```

Question 1: which goods category do customers spent more money on?

```python
categories_dictionary = {0:"Wines", 1:"Fruits", 2:"Meat products", 
                         3:"Fish products", 4:"Sweet products", 5:"Gold products"}

categories = ["MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", 
              "MntSweetProducts", "MntGoldProds"]

sums_dictionary = {0:[], 1:[], 2:[], 3:[], 4:[], 5:[]}

for index, category in enumerate(categories):
    current_sum = data[category].sum()
    sums_dictionary[index].append(current_sum)
    
max_amount = max(sums_dictionary.values())

for index in sums_dictionary:
    if sums_dictionary[index] == max_amount:
        print(str(categories_dictionary[index]) + ", " + str(sums_dictionary[index][0]))
```

Question 2: what is the year of birth of a person with the largest income?

```python
max_income = data["Income"].max()
data_max_income = data[data["Income"] == max_income][["Year_Birth", "Income"]].copy()
print(data_max_income)
```

Question 3: do people born in 1980 and 1990 have different income?

```python
years = [1980, 1990]

income_means = []
for year in years:
    data_current_year = data[data["Year_Birth"] == year].copy()
    mean_income = round(data_current_year["Income"].mean(), 2)
    income_means.append(mean_income)
    
for i in range(len(years)):
    print("Year, income mean: " + str(years[i]) + ", " + str(income_means[i]))
```
